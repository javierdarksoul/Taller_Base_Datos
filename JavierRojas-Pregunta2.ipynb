```python
import pandas as pd

lluvia = pd.read_csv("meteo.csv")
#lluvia = pd.get_dummies(lluvia)
lluvia[:20]
```

```python
def var_approx(cols):
    global prom
    Min = cols[0]
 
    if pd.isnull(Min):
        return prom
    else:
        return Min
```

```python
def date_val(cols):
    Min=cols[0]
    if Min=="No":
        return 0
    else:
        return 1
```

## limpieza de la data

```python
lluvia.isnull().sum()
```

```python
#dropeamos las variables "evaporation", "shunshine", "cloud9am" "cloud9pm" debido a que tienen demasiados datos nulos
lluvia = lluvia.drop(['Date','Evaporation','Sunshine','Cloud9am','Cloud3pm'], 1)
```

```python
#calculamos promedios para reemplazarlos
global prom

#para min temperatura
prom=lluvia['MinTemp'].sum()/142193
lluvia['MinTemp'] = lluvia[['MinTemp', 'Rainfall']].apply(var_approx, axis=1)

#para Max tempratura
prom=lluvia['MaxTemp'].sum()/142193
lluvia['MaxTemp'] = lluvia[['MaxTemp', 'Rainfall']].apply(var_approx, axis=1)

#para Rainfall
prom=lluvia['Rainfall'].sum()/142193
lluvia['Rainfall'] = lluvia[['Rainfall', 'MaxTemp']].apply(var_approx, axis=1)

#para WindGustSpeed
prom=lluvia['WindGustSpeed'].sum()/142193
lluvia['WindGustSpeed'] = lluvia[['WindGustSpeed', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['WindSpeed9am'].sum()/142193
lluvia['WindSpeed9am'] = lluvia[['WindSpeed9am', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['WindSpeed3pm'].sum()/142193
lluvia['WindSpeed3pm'] = lluvia[['WindSpeed3pm', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['Humidity9am'].sum()/142193
lluvia['Humidity9am'] = lluvia[['Humidity9am', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['Humidity3pm'].sum()/142193
lluvia['Humidity3pm'] = lluvia[['Humidity3pm', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['Pressure9am'].sum()/142193
lluvia['Pressure9am'] = lluvia[['Pressure9am', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['Pressure3pm'].sum()/142193
lluvia['Pressure3pm'] = lluvia[['Pressure3pm', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['Temp9am'].sum()/142193
lluvia['Temp9am'] = lluvia[['Temp9am', 'Rainfall']].apply(var_approx, axis=1)

prom=lluvia['Temp3pm'].sum()/142193
lluvia['Temp3pm'] = lluvia[['Temp3pm', 'Rainfall']].apply(var_approx, axis=1)

```

```python
lluvia['RainTomorrow'] = lluvia[['RainTomorrow', 'Rainfall']].apply(date_val, axis=1)
```

```python
#convertimos las columnas no binarias en binarias
lluvia=pd.get_dummies(lluvia)
```

```python
#selección de la primera columna de nuestro dataset (tamaño de la populacion)
X = lluvia.iloc[0:len(lluvia),0:12]
#selección de la segunda columna de nuestro dataset (volumen de negocio)
y = lluvia.iloc[0:len(lluvia),13]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=25)

from sklearn import tree
#Entrenamiento
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
#Predicciones
y_pred = clf.predict(X_test)
#Evaluación del rendimiento del clasificador
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
#Print de la matriz de confusión
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
##regresion logistica

from sklearn.linear_model import LogisticRegression

#Aprendizaje
LogReg = LogisticRegression()
LogReg.fit(X_train, y_train)
```

```python
#Predicciones
y_pred = LogReg.predict(X_test)
#Evaluación del rendimiento del clasificador
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
#Print de la matriz de confusión
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
##random forest
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=15)
clf = clf.fit(X_train, y_train)

#Predicciones
y_pred = clf.predict(X_test)
#Evaluación del rendimiento del clasificador
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
#Print de la matriz de confusión
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
